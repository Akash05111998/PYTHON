```python
# function for  enter correct date

import logging
from datetime import datetime

# Configure logging
logger = logging.getLogger()
logger.setLevel(logging.INFO)

# File handler
file_handler = logging.FileHandler('date_validation.log')
file_handler.setLevel(logging.INFO)
file_handler.setFormatter(logging.Formatter('%(asctime)s - %(levelname)s - %(message)s'))

# Console handler
console_handler = logging.StreamHandler()
console_handler.setLevel(logging.INFO)
console_handler.setFormatter(logging.Formatter('%(asctime)s - %(levelname)s - %(message)s'))

# Add handlers to the logger
logger.addHandler(file_handler)
logger.addHandler(console_handler)

def validate_date(date_str, date_format='%Y-%m-%d'):
    try:
        if date_str is None:
            logger.error("Validation failed: date cannot be None")
            raise ValueError("date cannot be None")
        elif not isinstance(date_str, str):
            logger.error("Validation failed: date must be a string")
            raise ValueError("date must be a string")
        
        # Try to parse the date
        datetime.strptime(date_str, date_format)
        logger.info(f"Validation successful: '{date_str}' is a valid date")
        return True
    except ValueError as e:
        logger.exception(f"Exception occurred: {e}")
        return False

def test_date_validation(dates, date_format='%Y-%m-%d'):
    for date_str in dates:
        if validate_date(date_str, date_format):
            print(f"'{date_str}' is valid")
        else:
            print(f"'{date_str}' is invalid")

# Example usage
if __name__ == "__main__":
    dates = input("Enter dates separated by spaces (format YYYY-MM-DD): ").strip().split()
    test_date_validation(dates)
```

```python
from logging_config import configure_logging
from datetime import datetime
import logging

# Ensure logging is configured

def validate_date(date_str):
    if not date_str or not isinstance(date_str, str):
        logging.error(f"Validation failed: {date_str} is not a valid string.")
        raise ValueError("Date must be a non-empty string.")
    
    try:
        date_obj = datetime.strptime(date_str, '%d-%m-%Y')
    except ValueError:
        try:
            date_obj = datetime.strptime(date_str, '%d/%m/%Y')
        except ValueError:
            logging.error(f"Validation failed: {date_str} is not a valid date format.")
            raise ValueError("Date format must be DD-MM-YYYY or DD/MM/YYYY.")
    
    formatted_date = date_obj.strftime('%d-%m-%Y')
    logging.info(f"Validation successful: '{date_str}' is a valid date, converted to '{formatted_date}'")
    return formatted_date

def test_date_validation(dates):
    for date_str in dates:
        try:
            formatted_date = validate_date(date_str)
            print(f"'{date_str}' is valid and converted to '{formatted_date}'")
        except ValueError as e:
            print(f"'{date_str}' is invalid: {e}")

# Example usage
dates = input("Enter date of birth (formats DD-MM-YYYY and DD/MM/YYYY): ").strip().split()
test_date_validation(dates)
```
